```python
import numpy as np
np.random.seed(42)

import pandas as pd
import matplotlib as plt

import statsmodels.api as sm
from statsmodels.tsa.api import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
myts = np.random.normal(size = 60)
dates = pd.date_range(
    start = '2012-01',
    end = '2017-01',
    freq = 'ME')
myts_series = pd.Series(
    myts,
     index = dates
)
myts_series.plot(title = 'Random Time Series')
```

## Using a csv

```python
import csv
```

```python
def read_csv(file_path):
    try:
        with open(file_path, mode = 'r', encoding = 'utf-8') as file:
            csv_reader = csv.reader(file)
            lines = list(csv_reader)
        return lines
    
    except FileNotFoundError:
        print(f'Error: File {file_path} not found.')
        return None

    except Exception as e:
        print(f'An error ocurred while reading the file: {file_path}')
        return None
```

```python
file_path = 'data/importar.csv'
```

```python
data = read_csv(file_path)
```

```python
values = [float(row[1]) for row in data]
```

```python
stValues = pd.period_range(start = '1884', end = '1939', freq = 'Y')
```

```python
time_series = pd.Series(values, index = stValues)
```

```python
time_series.plot(title = 'Random Time Series 2')
```
